```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
school_data_complete.head()
```

```python
# Total Number of Schools
total_num_schools = school_data_complete["school_name"].nunique()

# Total Number of Students
total_num_students = school_data_complete["Student ID"].nunique()

# Total Budget
total_budget = sum(school_data_complete["budget"].unique())

# Average Math Score
total_avg_math = round(school_data_complete["math_score"].mean(), 2)

# Average Reading Score
total_avg_read = round(school_data_complete["reading_score"].mean(), 2)

# Percent Students Passing Math 
passing_math = round(((sum(student >= 70 for student in school_data_complete["math_score"])
                       /total_num_students)*100),2)

# Percent Students Passing Reading
passing_read = round(((sum(student >= 70 for student in school_data_complete["reading_score"])
                       /total_num_students)*100),2)

# Percent Overall Passing Rate
overall_pass_rate = round((total_avg_math + total_avg_read)/2, 2)


```

```python
# Print District Summary Dataframe
district_sum = [
                {
             "Total Schools": total_num_schools,
             "Total Students": total_num_students,
             "Total Budget": total_budget,
             "Average Math Score": total_avg_math,      
             "Average Reading Score": total_avg_read,
             "% Passing Math": passing_math,
             "% Passing Reading": passing_read,
             "% Overall Score": overall_pass_rate,      
                }
              ]
district_sum_df = pd.DataFrame(district_sum)
district_sum_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Create base dataframe
school_sum_schools = school_data_complete[["school_name", "type", "student_name", "budget",
                                           "math_score", "reading_score"]]

# GroupBy State and Type
school_group = school_sum_schools.groupby(["school_name", "type"])

# Student totals
students=school_group["student_name"].count()

# Budget totals
budget = school_group["budget"].unique()

# Budget per Student
budget_per_student=budget/students

# Average Math Score
avg_math = round(school_group["math_score"].mean(), 2)

# Average Reading Score
avg_read = round(school_group["reading_score"].mean(), 2)

# Percent Students Passing Math 
math_pass=school_data_complete[school_data_complete['math_score'] >= 70].groupby(['school_name', "type"])["math_score"].count()
math_test=school_data_complete.groupby(['school_name', "type"])["math_score"].count()
math_percent_pass=round((math_pass/math_test)*100,2)

# Percent Students Passing Reading
read_pass=school_data_complete[school_data_complete['reading_score'] >= 70].groupby(['school_name', "type"])["reading_score"].count()
read_test=school_data_complete.groupby(['school_name', "type"])["reading_score"].count()
read_percent_pass=round((read_pass/read_test)*100,2)

# Percent Overall Passing Rate
all_pass_rate = round((read_percent_pass + math_percent_pass)/2,2)
```

```python
school_summary_table = pd.DataFrame({"Total Students": students,
                                    "Total Budget": budget,
                                    "Per Student Budget": budget_per_student,
                                    "Average Math Score":avg_math,
                                    "Average Reading Score":avg_read,
                                    "% Passing Math":math_percent_pass,
                                    "% Passing Reading":read_percent_pass,
                                    "% Overall Passing":all_pass_rate})
school_summary_table
```
